```python
import numpy as np
from sklearn.datasets import make_regression
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.wrappers.scikit_learn import KerasRegressor
X, y = make_regression(n_samples = 500, n_features = 2,n_informative = 2, noise = 5,random_state = 0)
```

```python
def model_build():
    mod = Sequential()
    mod.add(Dense(16, input_dim = 2, activation = "relu"))
    mod.add(Dense(8, activation = "relu"))
    mod.add(Dense(1))
    mod.compile(loss='mean_squared_error', optimizer='adam')
    return mod
```

```python
YourModel = KerasRegressor(build_fn = model_build,epochs = 400,batch_size = 10,verbose = 1)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
kf = KFold(n_splits = 5)
results = cross_val_score(YourModel, X, y, cv=kf)
```

```python
print("Final Cross Validation Loss:", abs(results.mean()))
```

```python
import pandas as pd
X = pd.read_csv("./Applied-Deep-Learning-with-Keras/Lesson02/bank_data_feats.csv", index_col = 0)
y = pd.read_csv("./Applied-Deep-Learning-with-Keras/Lesson02/bank_data_target.csv", index_col = 0)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
seed = 1
np.random.seed(seed)
model = Sequential()
model.add(Dense(8, input_dim = 32, activation = 'relu'))
model.add(Dense(8, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
model.compile(loss = 'binary_crossentropy',optimizer = 'sgd',metrics = ['accuracy'])
trainRes = model.fit(X_train, y_train,batch_size = 10,epochs = 400,validation_data=(X_test, y_test),verbose = 0)
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['figure.figsize'] = (6.0, 4.0)
plt.plot(trainRes.history['loss'])
plt.plot(trainRes.history['val_loss'])
plt.ylim(0, 1)
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train loss', 'test loss'], loc='upper right')

print("BA =", max(trainRes.history['val_acc']))
```

```python
from keras.regularizers import l2

model = Sequential()
model.add(Dense(32, input_dim = 32, activation = 'tanh',kernel_regularizer=l2(0.01)))
model.add(Dense(16, activation = 'relu'))
model.add(Dense(8, activation = 'tanh'))
model.add(Dense(4, activation = 'tanh'))
model.add(Dense(1, activation = 'sigmoid'))

model.compile(loss = 'binary_crossentropy',optimizer = 'sgd',metrics = ['accuracy'])
```

```python
trainRes = model.fit(X_train, y_train,batch_size = 10,epochs = 300,validation_data=(X_test, y_test),verbose = 0)
plt.plot(trainRes.history['loss'])
plt.plot(trainRes.history['val_loss'])
plt.ylim(0, 1)
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train loss', 'test loss'], loc='upper right')

print("BA =", max(trainRes.history['val_acc']))
```
