```python
dx = 0.0000001
start = 0.0
end = 3.0
loopLen = (end - start) / dx
print(loopLen)
x = 0
y = 0
area = 0
for i in range(int(loopLen)):
    y = x ** 2
    x += dx
    area += dx * y
print(str(area))
```

```python
def intersect(g, h):
    i = []
    for x in g:
        if x in h:
            i.append(x)
    return i
l1 = "asdqx"
l2 = "qwe"
l3 = "zxc"
print(intersect(l1, l2))
print(intersect(l1, l3))
```
